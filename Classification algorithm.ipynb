# Data Preprocessing

```python
# Encoding the target variable from categorical variable to numericalvariable
# Attrition being the target variable
dataset['Attrition'] = dataset['Attrition'].replace(['Yes', 'No'], [1, 0])
```

```python
# Initializing One-hot encoder for categorical variables to avoid multicollinearity
from sklearn import preprocessing
enc = preprocessing.OneHotEncoder(drop='first')
onehots = enc.fit_transform(dataset[['BusinessTravel', 'Department', 'Gender', 'EducationField', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']]).toarray()

# creating the new dataset
cols = []
for i in enc.categories_:
    i = np.delete(i,0)
    cols.extend(i)
dataset = dataset.join(pd.DataFrame(onehots, columns=cols))

dataset = dataset.drop(['BusinessTravel', 'Department', 'Gender', 'EducationField', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime'], axis=1)
dataset
```

```python
# Check correlation between features using a heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```

# Scaling Features

```python
# Splitting the dataset into features (X) and target (y)
from sklearn.model_selection import train_test_split
X = dataset.drop('Attrition', axis=1)
y = dataset['Attrition']
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

```python
from imblearn.over_sampling import SMOTE
# Apply SMOTE to balance the classes in the training set
resampler = SMOTE(random_state=0)
X_train_smote, y_train = resampler.fit_resample(X_train, y_train)
sns.countplot(x=y_train)
plt.show()
```

```python
print("\n After Smote" , y_train.value_counts())
```

# Feature Selection

```python
# Standardizing the features for classification
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.feature_selection import RFECV
sc = StandardScaler()
X_train_sm = sc.fit_transform(X_train_smote)
X_test = sc.transform(X_test)
```

```python
# Using RandomForestClassifier as the base model
rf = RandomForestClassifier(random_state=0) 
rfecv = RFECV(rf, cv=3, step=5)
X_train_sm = rfecv.fit_transform(X_train_sm, y_train)
X_test_sm = rfecv.transform(X_test)
print(f"Number of remaining features: {X_train_sm.shape[1]}")
```

```python
# Plotting the model of accuracy against the number of features
plt.figure(figsize=(15, 6))
plt.title('Number of Features Included vs Accuracy')
plt.xlabel('Number of Features Included')
plt.ylabel('Model Accuracy')
num_scores = len(rfecv.cv_results_['mean_test_score'])
plt.plot(np.linspace(0,40,num_scores), rfecv.cv_results_['mean_test_score'])
plt.show()
```

# Random Forest

```python
# Fitting Random Forest to the training set
rf_selectedfeatures = RandomForestClassifier()
rf_selectedfeatures.fit(X_train_sm, y_train)
```

```python
# Predicting the test data
y_pred = rf_selectedfeatures.predict(X_test_sm)
print(f"Accuracy Score: {accuracy_score(y_test,y_pred)*100:.2f}%")
cm = confusion_matrix(y_test,y_pred)
ax = sns.heatmap(cm, cmap='flare',annot=True, fmt='d')
plt.xlabel("Predicted Class",fontsize=12)
plt.ylabel("True Class",fontsize=12)
plt.title("Confusion Matrix",fontsize=12)
plt.show()
```

```python
# Evaluating the model
from sklearn import metrics
acc=metrics.accuracy_score(y_test,y_pred)
print('accuracy:%.2f\n\n'%(acc))
cm=metrics.confusion_matrix(y_test,y_pred)
print('Confusion Matrix:')
print(cm,'\n\n')
print('-------------------------------------------------------')
result=metrics.classification_report(y_test,y_pred)
print('Confusion Matrix:\n')
print(result)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
# Create a Logistic Regression model
logistic_regression = LogisticRegression()
# Fitting Logistic Rogression to the training set
logistic_regression.fit(X_train_sm, y_train)
```

```python
# Predicting the test data
y_pred = logistic_regression.predict(X_test_sm)
print(f"Accuracy Score: {accuracy_score(y_test,y_pred)*100:.2f}%")
cm = confusion_matrix(y_test,y_pred)
ax = sns.heatmap(cm, cmap='flare',annot=True, fmt='d')
plt.xlabel("Predicted Class",fontsize=12)
plt.ylabel("True Class",fontsize=12)
plt.title("Logistic Regression Confusion Matrix",fontsize=12)
plt.show()
```

```python
# Evaluating the model
from sklearn import metrics
acc=metrics.accuracy_score(y_test,y_pred)
print('accuracy:%.2f\n\n'%(acc))
cm=metrics.confusion_matrix(y_test,y_pred)
print('Confusion Matrix:')
print(cm,'\n\n')
print('-------------------------------------------------------')
result=metrics.classification_report(y_test,y_pred)
print('Confusion Matrix:\n')
print(result)
```
